# Analyse Imported Data

To import data and analyse it, we only need the `Track` class.

```python
from iscat_lib.analysis import Track
```

We can now import data from a `.csv` file. There are multiple datafiles in this directory; we only load one of them for now.

```python
track = Track.from_file("C:\\Users\\John\\Desktop\\Yashar Data\\data_10.csv", unit="micrometres")
track
```

To view the trajectory, we can use `Track.plot_trajectory()`.

```python
track.plot_trajectory()
```

We can choose the range of data used for the fits with the keyword argument `fractionFitPoints`.

```python
track.msd_analysis(fractionFitPoints = 0.15)
track.plot_msd_analysis_results()
```

```python
track.adc_analysis(fractionFitPoints = 0.15)
track.plot_adc_analysis_results()
```

```python
track.sd_analysis(fractionFitPoints = 0.15, maxfev=10000)
track.plot_sd_analysis_results()
```

# Working with Multiple Tracks

`ListOfTracks` can be used to analyse multiple tracks in bulk.

```python
from iscat_lib.analysis import Track
from iscat_lib.analysis import ListOfTracks
```

Load the tracks one by one.

```python
track1 = Track.from_file("track1.csv", unit="micrometres")
track2 = Track.from_file("track2.csv", unit="micrometres")
track3 = Track.from_file("track3.csv", unit="micrometres")
```

Construct a new `ListOfTracks` from the individual tracks.

```python
tracks = ListOfTracks([track1, track2, track3])
```

Plot of all tracks in `tracks`. Each track is centered so that its origin lies at (0, 0). The time is color-coded.

```python
tracks.plot_trajectories(cmap='plasma')
```

Analyse all tracks in bulk using `adc_analysis`.

```python
tracks.adc_analysis()
```

`smart_averaging` averages all tracks that are in the same category and returns the fraction of tracks in each catgegory. (Note: For now, this only works if all tracks have the same length.)

```python
tracks.summary(plot_dapp=True, plot_pie_chart=True)
```

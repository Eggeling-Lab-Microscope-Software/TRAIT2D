```python
%load_ext autoreload
%autoreload 2
from iscat_lib import simulators
import csv
```

```python
# Parameters for free diffusion
params = dict()
params["Tmax"] = 0.25
params["dt"] = 5e-4
params["L"] = 10
params["dL"] = 1e-3
params["d"] = 0.8
#params["seed"] = 41
params["quantize"] = False
```

```python
#fields like in the Imaris files

fields = ['x','y','z','unit','category','collection','birth','death','trackID','id','dontmatter1','dontmatter2','dontmatter3','dontmatter4','dontmatter5']
```

```python
i = 1000000009

sfree = simulators.BrownianDiffusion(**params,seed = i)
sfree.run()
sfree.display_trajectory()
t1 = sfree.trajectory
with open(r'masso.csv','a',newline = '') as outputfile:
    writer = csv.DictWriter(outputfile, fieldnames = fields)
    #writer.writeheader()
    for j in range(0,len(t1['x'])):
        writer.writerow({'x': t1['x'][j],'y': t1['y'][j],'z':0,'unit':'um','category' : 'spot','collection':'position','birth': t1['t'][j], 'death':(t1['t'][j]+params["dt"]), 'trackID' : i,'id':j,'dontmatter1': 0,'dontmatter2': 0,'dontmatter3': 0,'dontmatter4': 0,'dontmatter5': 0})

```

```python
#Parameters for hop diffusion
params = dict()
params["Tmax"] = 0.25
params["dt"] = 5e-4 
params["L"] = 3
params["dL"] = 1e-3
params["Df"] = 0.8
params['HL'] = 200e-3 # compartment size (um)
params['HP'] = 0.01 # hop probability
#params["seed"] = 43
params["quantize"] = False
```

```python
# Test HoppingDiffusion without quantize

#i = 1000000009
for i in range(1000000000,1000000010):
    shop = simulators.HoppingDiffusion(**params,seed = i)
    shop.run()
    shop.display_trajectory()
    t1 = shop.trajectory

    with open(r'hopdiffusion_simulation_CompSize=160nm_HopProb=0.01.csv','a',newline='') as outputfile:
        writer = csv.DictWriter(outputfile,fieldnames = fields)
        if i == 1000000000:
            writer.writeheader();writer.writeheader();writer.writeheader()
        for j in range(0,len(t1['x'])):
            writer.writerow({'x': t1['x'][j],'y': t1['y'][j],'z':0,'unit':'um','category' : 'spot','collection':'position','birth': t1['t'][j], 'death':(t1['t'][j]+params["dt"]), 'trackID' : i,'id':j,'dontmatter1': 0,'dontmatter2': 0,'dontmatter3': 0,'dontmatter4': 0,'dontmatter5': 0})
```
